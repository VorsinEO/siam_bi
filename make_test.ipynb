```python
def add_info(ver,
            public,
            local,
            desc):
    exps  = {
    'ver':[],
    'public':[],
    'local':[],
    'desc':[]
    }
    exps['ver'].append(ver)
    exps['public'].append(public)
    exps['local'].append(local)
    exps['desc'].append(desc)
    return pd.DataFrame(exps)
    
```
